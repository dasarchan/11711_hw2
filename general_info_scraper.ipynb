```python
import requests
from bs4 import BeautifulSoup
import json
import re
import pandas as pd
import time
import nltk
```

# webscraping

```python
pgh_wiki = "https://en.wikipedia.org/wiki/Pittsburgh"
pgh_history_wiki = "https://en.wikipedia.org/wiki/History_of_Pittsburgh"
cmu_about = "https://www.cmu.edu/about/"
cmu_history = "https://www.cmu.edu/about/history.html"
```

```python
header = {'Accept': 'text/html', 
            "User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
```

```python
links = [ ]
text_chunks = [ ]
```

```python
page_response = requests.get(pgh_wiki, headers = header, timeout = 20)
page_root = BeautifulSoup(page_response.content, 'html.parser')
text = page_root.get_text()
tt = nltk.tokenize.TextTilingTokenizer()
tiles = tt.tokenize(text)

for tile in tiles:
    text_chunks.append(tile.replace("\n", " ").replace("\t", " ").replace(u"\xa0", u" "))
    links.append(pgh_wiki)
```

```python
text_chunks = text_chunks[5:15] + text_chunks[18:78]
```

```python
links = links[:70]
```

```python
page_response = requests.get(pgh_history_wiki, headers = header, timeout = 20)
page_root = BeautifulSoup(page_response.content, 'html.parser')

text = page_root.get_text()
tt = nltk.tokenize.TextTilingTokenizer()
tiles = tt.tokenize(text)

pgh_history_text = [ ]
pgh_history_links = [ ]
for tile in tiles:
    pgh_history_text.append(tile.replace("\n", " ").replace("\t", " ").replace(u"\xa0", u" "))
    pgh_history_links.append(pgh_history_wiki)
```

```python
pgh_history_text = pgh_history_text[2:42]
pgh_history_links = pgh_history_links[:len(pgh_history_text)]
```

```python
links += pgh_history_links
text_chunks += pgh_history_text
```

```python
cmu_about_text = cmu_about_text[1:]
cmu_about_links = cmu_about_links[1:]
```

```python
links += cmu_about_links
text_chunks += cmu_about_text
```

```python
pgh_britannica = "https://www.britannica.com/place/Pittsburgh"
page_response = requests.get(pgh_britannica, headers = header, timeout = 20)
page_root = BeautifulSoup(page_response.content, 'html.parser')

text = page_root.get_text()
tt = nltk.tokenize.TextTilingTokenizer()
tiles = tt.tokenize(text)

pgh_britannica_text = [ ]
pgh_britannica_links = [ ]
for tile in tiles:
    pgh_britannica_text.append(tile.replace("\n", " ").replace("\t", " ").replace(u"\xa0", u" "))
    pgh_britannica_links.append(pgh_britannica)
```

```python
pgh_britannica_text = pgh_britannica_text[7:]
pgh_britannica_links = pgh_britannica_links[7:]
text_chunks += pgh_britannica_text
links += pgh_britannica_links
```

```python
city_pgh_root = "https://www.pittsburghpa.gov/Home"
page_response = requests.get(city_pgh_root, headers = header, timeout = 20)
page_root = BeautifulSoup(page_response.content, 'html.parser')

city_pgh_links = [ ]
for i in page_root.find_all("li", {"class": "second-level-item"}):
    for a in i.find_all("a", href=True):
        city_pgh_links.append(a.get("href"))
```

```python
city_pgh_text = [ ]
city_pgh_link_list = [ ]
for link in city_pgh_links:
    print(link)
    page_response = requests.get(link, headers = header, timeout = 20)
    page_root = BeautifulSoup(page_response.content, 'html.parser')
    text = page_root.get_text()
    
    tt = nltk.tokenize.TextTilingTokenizer()
    try:
        tiles = tt.tokenize(text)
    
        for tile in tiles:
            city_pgh_text.append(tile.replace("\n", " ").replace("\t", " ").replace(u"\xa0", u" "))
            city_pgh_link_list.append(link)
    
    except ValueError:
        continue
    
    time.sleep(0.5)
```

```python
text_chunks += city_pgh_text
links += city_pgh_link_list
```

```python
visit_pgh_root = "https://www.visitpittsburgh.com/"
page_response = requests.get(visit_pgh_root, headers = header, timeout = 20)
page_root = BeautifulSoup(page_response.content, 'html.parser')
```

```python
visit_pgh_links = [visit_pgh_root]
for a in page_root.find_all("a", {"class": "nav__link"}):
    link = a.get("href")
    if link[0] == "h":
        visit_pgh_links.append(link)
```

```python
add_links = [ ]
for link in visit_pgh_links:
    page_response = requests.get(link, headers = header, timeout = 20)
    page_root = BeautifulSoup(page_response.content, 'html.parser')
    for a in page_root.find_all("a"):
        href = a.get("href")
        if len(href) > 32 and href[:32] == "https://www.visitpittsburgh.com/":
            add_links.append(href)
```

```python
visit_pgh_links = list(set(visit_pgh_links + add_links))
```

```python
visit_pgh_text = [ ]
visit_pgh_links_list = [ ]
for link in visit_pgh_links:
    print(link)
    page_response = requests.get(link, headers = header, timeout = 20)
    page_root = BeautifulSoup(page_response.content, 'html.parser')
    text = page_root.get_text()
    
    tt = nltk.tokenize.TextTilingTokenizer()
    
    try:
        tiles = tt.tokenize(text)

        for tile in tiles:
            visit_pgh_text.append(tile.replace("\n", " ").replace("\t", " ").replace(u"\xa0", u" "))
            visit_pgh_links_list.append(link)
    
    except ValueError:
        continue
    
    time.sleep(0.5)
```

```python
text_chunks += visit_pgh_text
links += visit_pgh_links_list
```

# pdf files

```python
from PyPDF2 import PdfReader
```

```python
from os import listdir
from os.path import isfile, join
pdfs = [f for f in listdir("pdf_files/") if isfile(join("pdf_files/", f))]
pdfs = pdfs[:2] + pdfs[3:]
```

```python
pdf_names = [ ]
pdf_text = [ ]
for pdf in pdfs:
    print(pdf)
#     file_name = pdf[:-4]
    reader = PdfReader("pdf_files/" + pdf)
    for page in reader.pages:
        page_text = page.extract_text()
        tt = nltk.tokenize.TextTilingTokenizer()
        try:
            tiles = tt.tokenize(page_text)
            for tile in tiles:
                pdf_text.append(tile.replace("\n", " ").replace("\t", " ").replace(u"\xa0", u" "))
                pdf_names.append(pdf)
        
        except ValueError:
            pdf_text.append(page_text.replace("\n", " ").replace("\t", " ").replace(u"\xa0", u" "))
            pdf_names.append(pdf)

#     pdf_df = pd.DataFrame(text, columns = ["text"])
#     pdf_df.to_csv(f"data/{file_name}.csv", index=False)
```

```python
text_chunks += pdf_text
links += pdf_names
```

```python
df = pd.DataFrame({"link": links, "text": text_chunks})
df.to_csv("scraped_data/general_info.csv", index=False)
```
