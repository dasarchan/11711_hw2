# webscraping

```python
pgh_wiki = "https://en.wikipedia.org/wiki/Pittsburgh"
pgh_history_wiki = "https://en.wikipedia.org/wiki/History_of_Pittsburgh"
cmu_about = "https://www.cmu.edu/about/"
cmu_history = "https://www.cmu.edu/about/history.html"
```

```python
header = {'Accept': 'text/html', 
            "User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
```

```python
page_response = requests.get(pgh_wiki, headers = header, timeout = 20)
page_root = BeautifulSoup(page_response.content, 'html.parser')
text = page_root.get_text()
tt = nltk.tokenize.TextTilingTokenizer()
tiles = tt.tokenize(text)

pgh_wiki_text = [ ]
for tile in tiles:
    pgh_wiki_text.append(tile.replace("\n", " ").replace("\t", " ").replace(u"\xa0", u" "))
```

```python
pgh_wiki_df = pd.DataFrame(pgh_wiki_text, columns = ["text"])
pgh_wiki_df.to_csv("data/pgh_wikipedia.csv", index=False)
```

```python
city_pgh_root = "https://www.pittsburghpa.gov/Home"
page_response = requests.get(city_pgh_root, headers = header, timeout = 20)
page_root = BeautifulSoup(page_response.content, 'html.parser')

city_pgh_links = [ ]
for i in page_root.find_all("li", {"class": "second-level-item"}):
    for a in i.find_all("a", href=True):
        city_pgh_links.append(a.get("href"))
```

```python
city_pgh_text = [ ]
for link in city_pgh_links:
    print(link)
    page_response = requests.get(link, headers = header, timeout = 20)
    page_root = BeautifulSoup(page_response.content, 'html.parser')
    text = page_root.get_text()
    
    tt = nltk.tokenize.TextTilingTokenizer()
    tiles = tt.tokenize(text)
    
    for tile in tiles:
        city_pgh_text.append(tile.replace("\n", " ").replace("\t", " ").replace(u"\xa0", u" "))
    
    time.sleep(0.5)
```

```python
city_pgh_df = pd.DataFrame(city_pgh_text, columns = ["text"])
city_pgh_df.to_csv("data/city_of_pgh.csv", index=False)
```

```python
visit_pgh_root = "https://www.visitpittsburgh.com/"
page_response = requests.get(visit_pgh_root, headers = header, timeout = 20)
page_root = BeautifulSoup(page_response.content, 'html.parser')
```

```python
visit_pgh_links = [visit_pgh_root]
for a in page_root.find_all("a", {"class": "nav__link"}):
    link = a.get("href")
    if link[0] == "h":
        visit_pgh_links.append(link)
```

```python
add_links = [ ]
for link in visit_pgh_links:
    page_response = requests.get(link, headers = header, timeout = 20)
    page_root = BeautifulSoup(page_response.content, 'html.parser')
    for a in page_root.find_all("a"):
        href = a.get("href")
        if len(href) > 32 and href[:32] == "https://www.visitpittsburgh.com/":
            add_links.append(href)
```

```python
visit_pgh_links = list(set(visit_pgh_links + add_links))
```

```python
visit_pgh_text = [ ]
for link in visit_pgh_links:
    print(link)
    page_response = requests.get(link, headers = header, timeout = 20)
    page_root = BeautifulSoup(page_response.content, 'html.parser')
    text = page_root.get_text()
    
    tt = nltk.tokenize.TextTilingTokenizer()
    tiles = tt.tokenize(text)
    
    for tile in tiles:
        visit_pgh_text.append(tile.replace("\n", " ").replace("\t", " ").replace(u"\xa0", u" "))
    
    time.sleep(0.5)
```

```python
visit_pgh_df = pd.DataFrame(visit_pgh_text, columns = ["text"])
visit_pgh_df.to_csv("data/visit_pgh.csv", index=False)
```

# pdf files

```python
from PyPDF2 import PdfReader
```

```python
from os import listdir
from os.path import isfile, join
pdfs = [f for f in listdir("pdf_files/") if isfile(join("pdf_files/", f))]
pdfs = pdfs[:2] + pdfs[3:]
```

```python
for pdf in pdfs:
    print(pdf)
    file_name = pdf[:-4]
    reader = PdfReader("pdf_files/" + pdf)
    text = [ ]
    for page in reader.pages:
        page_text = page.extract_text()
        tt = nltk.tokenize.TextTilingTokenizer()
        try:
            tiles = tt.tokenize(page_text)
            for tile in tiles:
                text.append(tile.replace("\n", " ").replace("\t", " ").replace(u"\xa0", u" "))
        
        except ValueError:
            text.append(page_text.replace("\n", " ").replace("\t", " ").replace(u"\xa0", u" "))

    
    pdf_df = pd.DataFrame(text, columns = ["text"])
    pdf_df.to_csv(f"data/{file_name}.csv", index=False)
```
